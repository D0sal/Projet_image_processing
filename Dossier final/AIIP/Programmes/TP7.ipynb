```python
"""
Pour ce notebook entier, 
TP7 Image processing
Nathan Oliveira Da Silva ING3A
"""
import matplotlib.pyplot as plt
import cv2
import numpy as np
import math
```

```python
img = cv2.imread('../../Images/mandrill.png',0)

# filtres
Gx = np.array([
        [-1,0,1],
        [-1,0,1],
        [-1,0,1]
    ])
Gy = np.array([
        [-1,-1,-1],
        [0,0,0],
        [1,1,1]
    ])

    
# application des filtres
Ix = cv2.filter2D(img,-1,Gx) 
Iy = cv2.filter2D(img,-1,Gy)

l = 0.1
temp = Ix*Iy
R = cv2.GaussianBlur(temp,[1,1],0)-l*pow((cv2.GaussianBlur(pow(Ix,2),[1,1],0)+cv2.GaussianBlur(pow(Ix,2),[1,1],0)),2)

seuil = 0.1
result = R<=seuil

result2 = cv2.cornerHarris(img,2,3,0.07)

# maths
plt.subplot(1,1,1)
plt.imshow(result,cmap='gray')
plt.show()

# cv2
plt.subplot(1,1,1)
plt.imshow(result2,cmap='gray')
plt.show()

```
