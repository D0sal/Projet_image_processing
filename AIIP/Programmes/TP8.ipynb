```python
"""
Pour ce notebook entier, 
TP8 Image processing
Nathan Oliveira Da Silva ING3A
"""
import matplotlib.pyplot as plt
import cv2
import numpy as np
import math
import random
```

```python
img = cv2.imread('../../Images/lena.jpg',0)

plt.subplot(2, 2, 1)
plt.hist(img.ravel(), 256, [0, 255])
plt.subplot(2, 2, 2)
plt.imshow(img,cmap='gray')
plt.show()

def algo(img):
    S = 300
    newS = random.randint(0,255)

    while(abs(newS-S)>0.1):
        S = newS

        g1 =img[img>=S]
        g2 =img[img<S]

        S1 = g1.mean()
        S2 = g2.mean()

        newS = (S1+S2)/2
        
    return newS


# calcul du seuil optimal
S = algo(img)

g1 = img>S
g2 = img<=S

# a faire : separer l'image avec le seuil

print('la valeur du seuil optimal est :',S)
plt.subplot(2, 2, 1)
plt.imshow(g1,cmap='gray')
plt.subplot(2, 2, 2)
plt.imshow(g2,cmap='gray')
plt.show()

```
