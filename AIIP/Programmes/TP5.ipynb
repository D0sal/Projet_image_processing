```python
"""
Pour ce notebook entier, 
TP5 Image processing
Nathan Oliveira Da Silva ING3A
"""
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

```python
import math

# fonction creation image 1
def createImg1():
    img = np.zeros((500,500))
    for i in range(len(img)):
        for j in range(i+1):
            img[i][j]=1

    return img

# cfonction calcul des grandients
def calcGradients(img):
    # filtres
    Gx = np.array([
        [-1,0,1],
        [-1,0,1],
        [-1,0,1]
    ])
    Gy = np.array([
        [-1,-1,-1],
        [0,0,0],
        [1,1,1]
    ])

    
    # application des filtres
    Ix = cv2.filter2D(img,-1,Gx) 
    Iy = cv2.filter2D(img,-1,Gy)

    plt.subplot(1,2,1)
    plt.imshow(Ix,cmap="Greys_r")
    plt.title("Ix")
    plt.subplot(1,2,2)
    plt.imshow(Iy,cmap="Greys_r")
    plt.title("Iy")
    plt.show()

    return (Ix,Iy)

# fonction calcul norme du gradient
def calcNormeGradients(Ix,Iy):
    Ig = np.sqrt(Ix*Ix + Iy*Iy)
    plt.subplot(1,1,1)
    plt.imshow(Ig,cmap='gray')
    plt.show()

    return Ig

# fonction contour
def contour(Ig,seuil):
    print(Ig.mean())
    img = Ig<=seuil

    plt.subplot(1,1,1)
    plt.imshow(img,cmap='gray')
    plt.show()
    return img
# programme 

# img = createImg1()
img = cv2.imread('../../Images/lena.jpg',0)

# calcul des gradiants
gradients = calcGradients(img)
normeGradient = calcNormeGradients(gradients[0],gradients[1])

imgContour = contour(normeGradient,6.15)
```
