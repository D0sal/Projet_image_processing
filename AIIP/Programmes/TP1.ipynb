```python
"""
Pour ce notebook entier, 
TP1 Image processing
Nathan Oliveira Da Silva ING3A
"""
# Exercice 1
import matplotlib.pyplot as plt
import numpy as np

img = np.zeros((500,500))
for i in range(len(img)):
    for j in range(i+1):
        img[i][j]=1


print(img)
plt.imshow(img,cmap="Greys_r")
```

```python
# Exercice 2 b
# triangle blanc coin haut droite
img1 = np.ones((10,10))
for i in range(len(img1)):
    for j in range(i):
        img1[i][j]=0

# triangle blanc coin haut gauche
img2 = np.ones((10,10))
for i in range(len(img2)):
    for j in range(i):
        img2[i,len(img2)-j-1]=0

# rectangle base noire
img3 = np.zeros((10,20))

# premier étage
temp = np.concatenate((img1,img2),1)

# image finale
result = np.concatenate((temp,img3),0)
print(result)
plt.imshow(result,cmap="Greys_r")
```

```python
# Exercice 2 c
img1 = np.zeros((10,10))
for i in range(len(img1)):
    for j in range(i):
        img1[i][j]=1

img2 = np.zeros((10,10))
for i in range(len(img2)):
    for j in range(i):
        img2[i,len(img2)-j-1]=1

# rectangle base noire premier étage
img3 = np.zeros((10,20))

# deuxième étage
temp = np.concatenate((img2,img1),1)

# image finale
result = np.concatenate((img3,temp),0)
print(result)
plt.imshow(result,cmap="Greys_r")
```

```python
# Exercice 3 Table d'echec
img = np.zeros((100,100))
for i in range(len(img)):
    if (i==0) or (i%2==0):
        for j in range (0,len(img)-1,2):
            img[i][j] = 1
    else :
        for j in range (1,len(img),2):
            img[i][j] = 1
print(img)
plt.imshow(img,cmap="Greys_r")
```

```python
# Exercice 7 a représentation indexée
import matplotlib.colors as clr

rouge = [1,0,0]
vert = [0,1,0]
bleu = [0,0,1]
blanc = [1,1,1]

map = clr.ListedColormap([rouge,vert,bleu]) # on créé une nouvelle colormap

I = np.array(([2,3,1],[3,1,2],[1,2,3]))


plt.imshow(I,cmap = map)
plt.show()
```

```python
# Exercice 7 b représentation indexée nouvelles couleurs
rouge = [1,0,0] # 1
vert = [0,1,0] # 2
bleu = [0,0,1] # 3 
jaune = [1,1,0] # 4
violet = [1,0,1] # 5 
vertSombre = [0,0.5,0] # 6
orange = [1,0.5,0] # 7
blanc = [1,1,1] # 8

map = clr.ListedColormap([rouge,vert,bleu,jaune,violet,vertSombre,orange,blanc]) # on créé une nouvelle colormap
I = np.array(([4,3,1],[2,8,4],[5,6,7])) 

plt.imshow(I,cmap = map)
plt.show()
```
